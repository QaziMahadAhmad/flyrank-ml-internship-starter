## 2. The question: decision, action, cost of a wrong call

*What decision does your work improve? Who acts on it? What does a wrong recommendation cost?*
## Research Question

**Decision:** Which content pages should be reviewed first for content refresh?

**Unit of analysis:** One content page.

**Output:** A ranked list of pages ordered by their refresh priority.

**Action:** An SEO or content team reviews the highest-ranked pages and decides whether to refresh, expand, rewrite, monitor, or leave them unchanged.

**Cost of a wrong recommendation:** A false positive wastes reviewer time on pages that did not need attention. A false negative means an important declining page may be missed, potentially reducing future search performance.

Machine learning can help because manually reviewing thousands of pages is expensive. A ranking model can prioritize the pages that deserve human attention first while keeping a human reviewer in the decision loop.
